```python
#intro to matplotlib
#graphical representation diffrent charts in form and the all charts represent diffrent values
#BAR PLOT
import matplotlib.pyplot as pt
values=[23,97,12,40,33]
scenes=['part1','part2','part3','part4','part5']

pt.bar(scenes,values)
#for filling of axis
pt.xlabel('Parts of novel',fontsize=19)
pt.ylabel('Popularity',fontsize=19) #if want to increase the font size
pt.title('Popularity of Novel',fontsize=20)
#if i want colors
pt.bar(scenes,values,color='purple')
#if i want diff colors in bar so just create the  list
colors=['red','yellow','blue','grey','purple']
pt.bar(scenes,values,color=colors,edgecolor='black')
#jo pehlai du gi wo nichai ai ga jo bd ma wo oper
#show means print
pt.show()
```

```python
import pandas as ps
import matplotlib.pyplot as pt
import numpy as np

data=ps.read_csv('test_data_csv.csv')
df=ps.DataFrame(data)

df = df.drop(df.index[9:18])
df['NAME']=df['NAME'].replace(np.nan,'kABEER')
df['SALARY']=df['SALARY'].replace(np.nan,1152)
print(df.drop_duplicates('NAME'))


pt.bar(df['NAME'],df['SALARY'])
pt.xlabel('Person Name',fontsize=19)
pt.ylabel('Salary',fontsize=19) #if want to increase the font size
pt.title('Details of Salary',fontsize=20)

pt.show()
```

```python
import matplotlib.pyplot as pt
import pandas as ps
import numpy as np

data=ps.read_csv('test_data_csv.csv')
df=ps.DataFrame(data)

df = df.drop(df.index[9:18])
df['SALARY']=df['SALARY'].replace(np.nan,1152)

group=df.groupby('GENDER').agg({'SALARY':'sum'})
pt.step(group.index,group['SALARY'],where='pre',marker='o')
pt.show()
```
